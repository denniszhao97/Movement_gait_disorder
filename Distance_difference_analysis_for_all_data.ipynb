```python
#Author Zhonghao(Dennis) Zhao
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
# Read, clean, and sort data files
# list for all types strides' mean and std, will be update on each set of data for all health people

health_exp_number = [300,301,302,303,304,305,306,307,308,309,310,311,312,313,314,318,320,321,322,323]
health_trail_number = [1,2]
total_lengthh = len(health_exp_number) * len(health_trail_number)
health_LL_mean = np.zeros(total_lengthh)
health_LL_std = np.zeros(total_lengthh)
health_RR_mean = np.zeros(total_lengthh)
health_RR_std = np.zeros(total_lengthh)
health_LR_mean = np.zeros(total_lengthh)
health_LR_std = np.zeros(total_lengthh)
health_RL_mean = np.zeros(total_lengthh)
health_RL_std = np.zeros(total_lengthh)
index = 0
for exp_n in health_exp_number:
    for trail_n in health_trail_number:
        dataset = pd.read_csv("DATA/"+str(exp_n)+"_B3_TRIAL0"+str(trail_n)+"_GAITCYCLES.csv")
        valid_data = dataset[dataset.Valid == True] #use valid == true data only
        Tor_data = valid_data[valid_data.EventType == "TOR"] # Get all tor data
        Tol_data = valid_data[valid_data.EventType == "TOL"] # Get all tol data
        Midl_data = valid_data[valid_data.EventType == "MidSSL"] # Get all midssl data
        Midr_data = valid_data[valid_data.EventType == "MidSSR"] # Get all midssr data
        Hsl_data = valid_data[valid_data.EventType == "HSL"] # Get all hsl data
        Hsr_data = valid_data[valid_data.EventType == "HSR"] # Get all hsr data
         
        # distance between two adjacent left steps
        Midl_xList = [element.X for element in Midl_data.itertuples()]
        Midl_yList = [element.Y for element in Midl_data.itertuples()]
        Midl_xarray = np.array(Midl_xList)#x coordinates
        Midl_yarray = np.array(Midl_yList)#y coordinates
        lengthl = min(len(Midl_xarray),len(Midl_yarray))
        Midl_difference = np.zeros(lengthl-1)
        for i in range(0,lengthl-1): #euclidean distance between two points
            Midl_difference[i] = math.sqrt((Midl_xarray[i+1]-Midl_xarray[i])**2+(Midl_yarray[i+1]-Midl_yarray[i])**2) 
        health_LL_mean[index] = Midl_difference.mean()
        health_LL_std[index] = Midl_difference.mean()
    
        # distance between two adjacent right steps
        Midr_xList = [element.X for element in Midr_data.itertuples()]
        Midr_yList = [element.Y for element in Midr_data.itertuples()]
        Midr_xarray = np.array(Midr_xList)#x coordinates
        Midr_yarray = np.array(Midr_yList)#y coordinates
        lengthr = min(len(Midr_xarray),len(Midr_yarray))
        Midr_difference = np.zeros(lengthr-1)
        for i in range(0,lengthr-1):
            Midr_difference[i] = math.sqrt((Midr_xarray[i+1]-Midr_xarray[i])**2+(Midr_yarray[i+1]-Midr_yarray[i])**2)
        health_RR_mean[index] = Midr_difference.mean()
        health_RR_std[index] = Midr_difference.std()
        
        # distance between left to right adjacent steps
        length = min(len(Midr_xarray), len(Midl_xarray)) # calculate result length
        Midlr_difference = np.zeros(length)
        for i in range(0,length):
            Midlr_difference[i] = math.sqrt((Midr_xarray[i]-Midl_xarray[i])**2+(Midr_yarray[i]-Midl_yarray[i])**2)
        health_LR_mean[index] = Midlr_difference.mean()
        health_LR_std[index] = Midlr_difference.std()
        
        # distance between right to left adjacent steps
        Midrl_difference = np.zeros(length-1)
        for i in range(0,length-1):
            Midrl_difference[i] = math.sqrt((Midl_xarray[i+1]-Midr_xarray[i])**2+(Midl_yarray[i+1]-Midr_yarray[i])**2)
        health_RL_mean[index] = Midrl_difference.mean()
        health_RL_std[index] = Midrl_difference.std()
        
        index += 1
```
